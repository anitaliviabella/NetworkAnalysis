6. **Horatio, Marcellus, and Bernardo:** They describe the ghostly apparition they witnessed on the previous nights to Hamlet.

7. **Hamlet and Ghost:** Hamlet encounters the ghost of his father, who reveals that he was murdered by Claudius and urges Hamlet to seek revenge.
